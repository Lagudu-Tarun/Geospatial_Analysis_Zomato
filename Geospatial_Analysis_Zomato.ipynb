Geospatial Analysis

```python
locations = pd.DataFrame({'Name':data['location'].unique()})
locations.head()
```

```python
!pip install geopy
```

```python
#Nominatim uses OpenStreetMap data to find locations on Earth by name and address (geocoding).
#It can also do the reverse, find an address for any location on the planet.

from geopy.geocoders import Nominatim
```

```python
#User_Agent is an http request header that is sent with each request.
#Nominatim requires this value to be set to your application name.
#The goal is to be able to limit the number of requests per application.
#timeout=None to make sure it must be connected to stable Internet Connection
geolocator = Nominatim(user_agent = 'app', timeout=None)
```

```python
lat_long = []
for location in locations['Name']:
    #To return lattitudes & longitudes
    location = geolocator.geocode(location)
    if location is None:
        lat_long.append(np.nan)
    else:
        geo = (location.latitude,location.longitude)
        lat_long.append(geo)
```

```python
locations['geo_loc'] = lat_long
```

```python
locations.head()
```

```python
locations.shape
```

```python
data['location'].value_counts()
```

```python
Reset_locations = pd.DataFrame(data['location'].value_counts().reset_index())
Reset_locations
```

```python
Reset_locations.columns = ['Name','count']
Reset_locations.head()
```

```python
locations.head()
```

```python
Restaurant_locations = Reset_locations.merge(locations,on='Name',how='left').dropna()
Restaurant_locations.head()
```

```python
#To convert column into an array

np.array(Restaurant_locations['geo_loc'])
```

```python
#Split the latitude and longitude

lat,lon = zip(*np.array(Restaurant_locations['geo_loc']))
```

```python
type(lat)
```

```python
Restaurant_locations['lat'] = lat
Restaurant_locations['lon'] = lon
```

```python
Restaurant_locations.head()
```

```python
Restaurant_locations.drop('geo_loc',axis=1,inplace=True)
```

```python
import folium 
from folium.plugins import HeatMap
```

```python
#Seting default location as BTM
def generateBaseMap(default_location=[12.97, 77.59], default_zoom_start=12):
    base_map = folium.Map(location=default_location, zoom_start=default_zoom_start)
    return base_map
```

```python
basemap = generateBaseMap()
```

```python
basemap
```

```python
Restaurant_locations.head()
```

```python
#convert into array using .value and to list using
values_list = Restaurant_locations[['lat','lon','count']].values.tolist()
values_list
```

```python
HeatMap(values_list,zoom = 20,radius = 15).add_to(basemap)
```
